'''

    在Javascript的世界中是沒有多執行緒的概念的，併發使用過使用事件驅動的方式來進行的， 所有的JavaScript程式都執行在一個執行緒中。
    在HTML5中引入web worker可以併發的處理任務，但沒有改變Javascript單執行緒的限制。
