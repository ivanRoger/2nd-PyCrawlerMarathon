### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'

'''
Your Code Here
'''
r = requests.get(url)
# 想要爬資料的目標網址
response = r.text
# 模擬發送請求的動作

r.encoding = 'utf-8'
print(r.text[0:3000])
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')

'''
Your Code Here
'''
# 想讓輸出結果換行可加入 +'\n'
html_doc = response
print(html_doc+'\n')
print(type(html_doc))
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ => ')

'''
Your Code Here
'''
soup = BeautifulSoup(html_doc, parser="html5lib")
print(soup)
print(type(soup))
```

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
url = 'https://www.zhihu.com/explore'
'''
Your Code Here
'''
headers = {'user-agent': 'user-agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'}

r = requests.get(url, headers = headers)
response = r.text

r.encoding = 'utf-8'
print(r.text[0:600])
```

```python
html_doc = response
print(html_doc+'\n')
print(type(html_doc))
```
