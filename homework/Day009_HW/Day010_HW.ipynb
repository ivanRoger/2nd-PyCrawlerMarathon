## Grab + PyQuery

```python
'''
Your Code
'''
from grab import Grab
g = Grab()
resp = g.go('https://www.youtube.com/')
resp.body
```

```python
from pyquery import PyQuery as pq
doc = pq(resp.body)
title = doc('title')
print(type(title), title.text())
```
